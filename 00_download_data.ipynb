# Download Daily & Weekly Price Data
**Assets:** SPY · WTI Crude Oil · Gold · VIX · OVX · DXY  
**Source:** Yahoo Finance via `yfinance`

```python
import os
import yfinance as yf
import pandas as pd

TICKERS = {
    "SPY": "SPY",        # S&P 500 ETF
    "OIL": "CL=F",      # WTI Crude Oil Futures
    "GOLD": "GC=F",     # Gold Futures
    "VIX": "^VIX",      # CBOE S&P 500 Volatility Index
    #"OVX": "^OVX",      # CBOE Crude Oil Volatility Index
    "DXY": "DX-Y.NYB",  # US Dollar Index
}

START = "2000-01-01"
DAILY_DIR = "data/daily"
WEEKLY_DIR = "data/weekly"
os.makedirs(DAILY_DIR, exist_ok=True)
os.makedirs(WEEKLY_DIR, exist_ok=True)
```

## Daily Data

```python
daily = {}
for name, ticker in TICKERS.items():
    df = yf.download(ticker, start=START, interval="1d", auto_adjust=True, progress=False)
    df.to_csv(os.path.join(DAILY_DIR, f"{name}_daily.csv"))
    daily[name] = df
    print(f"{name:5s} | {df.shape[0]:>6,} rows | {df.index.min().date()} → {df.index.max().date()}")
```

## Weekly Data

```python
weekly = {}
for name, ticker in TICKERS.items():
    df = yf.download(ticker, start=START, interval="1wk", auto_adjust=True, progress=False)
    df.to_csv(os.path.join(WEEKLY_DIR, f"{name}_weekly.csv"))
    weekly[name] = df
    print(f"{name:5s} | {df.shape[0]:>5,} rows | {df.index.min().date()} → {df.index.max().date()}")
```

## Quick Sanity Check

```python
for name in TICKERS:
    print(f"\n--- {name} (daily, last 5 rows) ---")
    display(daily[name].tail())
```

```python
print("Saved files:")
for d in [DAILY_DIR, WEEKLY_DIR]:
    for f in sorted(os.listdir(d)):
        size = os.path.getsize(os.path.join(d, f))
        print(f"  {d}/{f}  ({size/1024:.0f} KB)")
```
